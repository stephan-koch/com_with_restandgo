## login to Common Cloud Platform

```go
// Autogenerated by: https://mholt.github.io/json-to-go/ or https://transform.tools/json-to-go 

type Authjson struct {
  AccessToken string `json:"access_token"`
  TokenType   string `json:"token_type"`
  ExpiresIn   int    `json:"expires_in"`
}
```

```go

func login() string {

  apiUrl := "https://sso.common.cloud.hpe.com"
  resource := "/as/token.oauth2"  
  u, _ := url.ParseRequestURI(apiUrl)
  u.Path = resource
  urlStr := u.String()  
    
  data := url.Values{}
  
  data.Set("client_id","xxxxxxxx-xxxx-xxxx-xxxx-xxxxxxxxxxxx")  // obtain from common cloud console
  data.Set("client_secret","xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")  // obtain from common cloud console
  data.Set("grant_type","client_credentials")    
    
  client := &http.Client{}
  req, _ := http.NewRequest(http.MethodPost, urlStr, strings.NewReader(data.Encode())) // URL-encoded payload
    

  req.Header.Add("Content-Type", "application/x-www-form-urlencoded")   
    
  res, err := client.Do(req)  
    
  if err != nil {
   log.Fatalln(err)
  }  
    
  defer res.Body.Close()
  body, _ := ioutil.ReadAll(res.Body)
   
  
  fmt.Println("HTTP Response Code " + res.Status)
  //fmt.Println(string(body))
    
   
  var auth = &Authjson{}
  err := json.Unmarshal([]byte(body), auth)  
  if err != nil {
    fmt.Println("error:", err)
  }  
    
  //fmt.Println(auth.AccessToken)   
  return   auth.AccessToken
}


token := login()
   
fmt.Sprintf("My Bearer Token for 2 hours is: %s", token)   
```

```go
type postfwup struct {
  JobTemplateUri string `json:"jobTemplateUri"`
  ResourceUri string `json:"resourceUri"`  
    Data struct {
      Bundle_id   string `json:"bundle_id"`
    } `json:"data"`
}
```

```go
func dofwupdate(token string, jobrURI string, serverURI string, bundleURI string ) {
  reqUrl := "https://eu-central1-api.compute.cloud.hpe.com/compute-ops/v1beta1/jobs"
  var body = &postfwup{}
  body.JobTemplateUri = jobrURI
  body.ResourceUri = serverURI
  body.Data.Bundle_id = bundleURI  
  bodyjson, err := json.Marshal(body)  
  if err != nil {
    log.Fatalf("could not marshal JSON: %s", err)
  }
  
  req, _ := http.NewRequest("POST", reqUrl, bytes.NewBuffer(bodyjson))
  req.Header.Add("Content-Type", "application/json")
  req.Header.Add("Authorization", "Bearer " + token)
  res, _ := http.DefaultClient.Do(req)
  defer res.Body.Close()
  body, _ := ioutil.ReadAll(res.Body)

  fmt.Println(res.Status)
  fmt.Println(string(body))
}

dofwupdate(token, jobrURI, "/compute-ops/v1beta1/servers/867960-B21+CZJ84502PX", "/compute-ops/v1beta1/firmware-bundles/1523971143f67a278da840a2f798ba1d" )
```
